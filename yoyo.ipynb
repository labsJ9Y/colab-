```python
# CARD 
# SUIT(heart,dimond,...),RANK(2,3,...,jack,queen),VALUE(int value jack(11) and queen(12))
import random 
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')    # its a tuple
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card:

  def __init__(self,suit,rank): # no value, cuz we'll make the dictionary later for just value
      self.suit=suit
      self.rank=rank    # suit and rank are strings 
      self.value=values[rank]  # wrote rank to this value cuz we have already wrote rank in our instantanious shit thats why---(self,suit,rank)

  def __str__(self):
      return self.rank + " of " + self.suit
        
```

```python
class Deck:

  def __init__(self):
      self.all_cards = []
      for suit in suits:
          for rank in ranks:
              create_card = Card(suit,rank)
              self.all_cards.append(Card(suit,rank))


  def shuffle(self):

    random.shuffle(self.all_cards)


  def deal_one(self):                   # to grab one card
    return self.all_cards.pop()           # to return single card....becuz all_card is a list so we can use pop()   
```

```python
class Player:

  def __init__(self,name):    # each player has a name
      
      self.name = name
      self.all_cards = []    #each new player would have an empty hand....starting with an empty list

   # now we need to makes classes for to add and remove the card by the player to this all_card list which is their current hand of card


  def remove_one(self):
      return self.all_cards.pop(0)

  def add_cards(self,new_cards):
      # List of multiple card object 

      if type(new_cards) == type([]):           #--------- didnt understand this line, why there's a empty list with a "type " written there
        self.all_cards.extend(new_cards)
      # for a single card object 

      else:                                          # if we dont have a list of card and we had a single card 
          self.all_cards.append(new_cards)
              
  def __str__(self):
      return f'Player {self.name} has {len(self.all_cards)} cards.' 
```
